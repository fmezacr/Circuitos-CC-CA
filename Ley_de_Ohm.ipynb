# **Tutorial: Ley de Ohm en Circuitos de CC y CA**


## **1. Introducción**

La Ley de Ohm es una de las leyes fundamentales en el análisis de circuitos eléctricos. Se expresa matemáticamente como:

$$V = IR$$

Donde:
- \(V\) es el voltaje en volts (V).
- \(I\) es la corriente en amperios (A).
- \(R\) es la resistencia en ohmios (Ω).

Esta relación es válida para circuitos de corriente continua (CC) y tiene su extensión en circuitos de corriente alterna (CA), donde intervienen impedancias en lugar de resistencias puras.

---

## **2. Aplicación en Circuitos de Corriente Continua (CC)**

En circuitos de CC, la Ley de Ohm se aplica directamente con resistencias fijas. Veamos un ejemplo práctico:

### **Ejemplo 1: Circuito de CC**

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros del circuito
R = 10  # Resistencia en ohmios
V = np.linspace(0, 20, 100)  # Voltaje de 0V a 20V
I = V / R  # Aplicando la Ley de Ohm

# Graficar la relación V-I
plt.figure(figsize=(8, 5))
plt.plot(V, I, label='I = V / R', color='b')
plt.xlabel('Voltaje (V)')
plt.ylabel('Corriente (A)')
plt.title('Relación Voltaje-Corriente en CC')
plt.grid(True)
plt.legend()
plt.show()
```

En este gráfico se muestra cómo la corriente varía linealmente con el voltaje en un circuito de CC con una resistencia fija.

---

## **3. Aplicación en Circuitos de Corriente Alterna (CA)**

En circuitos de CA, la Ley de Ohm se generaliza usando **impedancias** en lugar de resistencias puras. La impedancia \(Z\) se define como:

$$V = IZ$$

Donde \(Z\) es la impedancia en ohmios (Ω), que puede incluir componentes resistivos, capacitivos e inductivos:

$$Z_R = R$$
$$Z_L = j\omega L$$
$$Z_C = \frac{1}{j\omega C}$$

### **Ejemplo 2: Circuito de CA con Impedancia**

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros del circuito
R = 8  # Resistencia en ohmios
L = 0.1 # Inductancia en Henrios
C = 100e-6  # Capacitancia en Faradios
f = 50  # Frecuencia en Hz
w = 2 * np.pi * f  # Frecuencia angular

# Impedancias en CA
Z_R = R
Z_L = 1j * w * L
Z_C = 1 / (1j * w * C)

# Fuente de voltaje en CA
t = np.linspace(0, 0.1, 1000)  # Tiempo en segundos
V_t = np.cos(w * t)  # Voltaje de la fuente en función del tiempo

# Corrientes a través de cada elemento
I_R = V_t / Z_R
I_L = V_t / np.abs(Z_L) * np.sin(w * t - np.pi / 2)  # Desfase -90° en L
I_C = V_t / np.abs(Z_C) * np.sin(w * t + np.pi / 2)  # Desfase +90° en C

# Graficar las señales
plt.figure(figsize=(10, 6))
plt.plot(t, V_t, label='Voltaje V(t)', color='black', linestyle='-')
plt.plot(t, I_R, label='Corriente en R', color='blue', linestyle='--')
plt.plot(t, I_L, label='Corriente en L', color='red', linestyle='-.')
plt.plot(t, I_C, label='Corriente en C', color='green', linestyle=':')

plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.title('Voltaje y Corriente en Circuitos de CA')
plt.legend()
plt.grid(True)
plt.show()


```

Este código ilustra cómo la Ley de Ohm en CA involucra el uso de números complejos para representar fase y magnitud de voltajes y corrientes.

---


## **4. Conclusión**

- En **CC**, la Ley de Ohm se aplica directamente con resistencias.
- En **CA**, la Ley de Ohm se generaliza usando impedancias, lo que introduce efectos de desfase.
- El análisis en CA es crucial para circuitos eléctricos que incluyen inductancias y capacitancias, como en la distribución de energía y sistemas electrónicos avanzados.
