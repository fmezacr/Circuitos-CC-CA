Este tutorial lo guiará a través de la teoría de los circuitos **RLC** y cómo se comportan bajo tres tipos de amortiguamiento: **subamortiguado**, **críticamente amortiguado**, y **sobreamortiguado**. También incluimos un código en Python para graficar las respuestas de cada caso y comparar las diferencias entre ellos. Recuerde que corresponde al caso de respuesta natural sin fuentes.

## Introducción a los Circuitos RLC

Un circuito **RLC** está compuesto por tres componentes principales:
- **R**: Resistencia.
- **L**: Inductancia.
- **C**: Capacitancia.

### Ecuación Diferencial en Circuitos RLC

La forma de la ecuación diferencial dependerá de si estamos analizando un **circuito RLC en serie** o **en paralelo**.

#### **Circuito RLC en Serie**

En un **circuito RLC en serie**, la **corriente** es la misma en todos los componentes, y la ecuación diferencial describe cómo la corriente $ i(t) $ varía en el tiempo. La ecuación diferencial es:

$
L \frac{d^2i(t)}{dt^2} + R \frac{di(t)}{dt} + \frac{1}{C} i(t) = V(t)
$

Donde:
- $ i(t) $ es la **corriente** que fluye por el circuito en el tiempo.
- $ V(t) $ es la **fuente de voltaje** aplicada al circuito.
- $ L $ es la inductancia (en henrios, H),
- $ R $ es la resistencia (en ohmios, Ω),
- $ C $ es la capacitancia (en faradios, F).

Si estamos considerando la **respuesta natural**, asumimos que $ V(t) = 0 $, ya que no hay una fuente externa, y estamos viendo cómo el sistema disipa la energía almacenada en los componentes. En este caso, la ecuación sería:

$
L \frac{d^2i(t)}{dt^2} + R \frac{di(t)}{dt} + \frac{1}{C} i(t) = 0
$

#### **Circuito RLC en Paralelo**

En un **circuito RLC en paralelo**, los elementos comparten el mismo voltaje, y la ecuación diferencial describe cómo cambia el **voltaje** $ v(t) $ en el tiempo. La ecuación diferencial es:

$
C \frac{d^2v(t)}{dt^2} + \frac{1}{R} \frac{dv(t)}{dt} + \frac{1}{L} v(t) = I(t)
$

Donde:
- $ v(t) $ es el **voltaje** en los componentes en paralelo.
- $ I(t) $ es la **fuente de corriente** aplicada al circuito.
- $ L $, $ R $ y $ C $ son las mismas como antes.

Si estamos analizando la **respuesta natural** del sistema, también asumimos que $ I(t) = 0 $, lo que nos da la siguiente ecuación:

$
C \frac{d^2v(t)}{dt^2} + \frac{1}{R} \frac{dv(t)}{dt} + \frac{1}{L} v(t) = 0
$

La solución de esta ecuación depende de dos parámetros importantes:
- **$ \alpha $**: El coeficiente de amortiguamiento, definido como $ \alpha = \frac{R}{2L} $.
- **$ \omega_0 $**: La frecuencia natural no amortiguada del sistema, definida como $ \omega_0 = \frac{1}{\sqrt{LC}} $.

El comportamiento del circuito **RLC** depende de la relación entre $ \alpha $ y $ \omega_0 $.


## Tipo de Respuesta en el Circuito RLC

En este análisis estamos considerando la **respuesta natural** del circuito **RLC**, lo que significa que no hay fuentes externas involucradas (como $ V(t) $ o $ I(t) $). La respuesta natural refleja cómo el circuito disipa la energía inicial almacenada en el inductor o en el capacitor, dependiendo de las condiciones iniciales. En un **circuito RLC en serie**, la variable que estamos graficando puede interpretarse como la **corriente** que fluye a través de todos los componentes, mientras que en un **circuito RLC en paralelo**, esta variable puede considerarse como el **voltaje** en los terminales comunes de los elementos. En ambos casos, estamos observando el comportamiento de la energía del sistema en función del tiempo para los tres tipos de amortiguamiento: **subamortiguado**, **críticamente amortiguado**, y **sobreamortiguado**. Este comportamiento es clave para entender cómo los circuitos RLC responden sin fuentes externas forzándolos.


## Casos de Amortiguamiento

Dependiendo del valor de $ \alpha $ en relación con $ \omega_0 $, existen tres posibles tipos de amortiguamiento en un circuito **RLC**:

### 1. Subamortiguado ($ \alpha < \omega_0 $)

Este caso ocurre cuando la resistencia es pequeña, lo que permite que el sistema tenga oscilaciones amortiguadas. La solución para la respuesta temporal del circuito es:

$
v(t) = e^{-\alpha t} \left( \cos(\omega_d t) + \frac{\alpha}{\omega_d} \sin(\omega_d t) \right)
$

Donde:
$ \omega_d = \sqrt{\omega_0^2 - \alpha^2} $ es la frecuencia angular amortiguada.

En este caso, el sistema oscila con una amplitud decreciente a medida que pasa el tiempo.

### 2. Críticamente Amortiguado ($ \alpha = \omega_0 $)

El caso críticamente amortiguado ocurre cuando la resistencia es exactamente suficiente para evitar las oscilaciones, pero el sistema aún retorna a su estado de equilibrio de manera rápida. La solución para este caso es:

$
v(t) = (1 + \alpha t) e^{-\alpha t}
$

En este escenario, el sistema decae rápidamente sin oscilaciones, y es el caso donde el sistema vuelve más rápido al equilibrio sin sobrepasarlo.

### 3. Sobreamortiguado ($ \alpha > \omega_0 $)

Cuando la resistencia es grande, el sistema es sobreamortiguado. En este caso, no hay oscilaciones y la respuesta del sistema es más lenta que en el caso críticamente amortiguado. La solución es:

$
v(t) = e^{s_1 t} + e^{s_2 t}
$

Donde:

$ s_1 = -\alpha + \sqrt{\alpha^2 - \omega_0^2} $

$ s_2 = -\alpha - \sqrt{\alpha^2 - \omega_0^2} $

En este caso, el sistema regresa lentamente al equilibrio, pero sin oscilaciones.

Observemos los 3 comportamientos:



```python
import numpy as np
import matplotlib.pyplot as plt

# Definir las constantes ajustadas para los tres casos
alpha_sub = 0.2  # Subamortiguado (alfa < omega_0)
alpha_crit = 1.5  # Criticamente amortiguado (alfa = omega_0)
alpha_over = 3.0  # Sobreamortiguado (alfa > omega_0)
omega_0 = 1.5  # Frecuencia natural no amortiguada para todos los casos

# Definir la función para los diferentes tipos de amortiguamiento
def subamortiguado(t, alpha, omega_0):
    # Solución para el caso subamortiguado
    omega_d = np.sqrt(omega_0**2 - alpha**2)
    return np.exp(-alpha * t) * (np.cos(omega_d * t) + (alpha / omega_d) * np.sin(omega_d * t))

def criticamente_amortiguado(t, alpha, omega_0):
    # Solución para el caso críticamente amortiguado
    return (1 + alpha * t) * np.exp(-alpha * t)

def sobreamortiguado(t, alpha, omega_0):
    # Solución para el caso sobreamortiguado
    s1 = -alpha + np.sqrt(alpha**2 - omega_0**2)
    s2 = -alpha - np.sqrt(alpha**2 - omega_0**2)
    return np.exp(s1 * t) + np.exp(s2 * t)

# Generar el tiempo para la gráfica
t = np.linspace(0, 20, 1000)

# Calcular las respuestas para los tres casos
response_sub = subamortiguado(t, alpha_sub, omega_0)
response_crit = criticamente_amortiguado(t, alpha_crit, omega_0)
response_over = sobreamortiguado(t, alpha_over, omega_0)

# Graficar las respuestas
plt.figure(figsize=(10, 6))
plt.plot(t, response_sub, label='Subamortiguado (α < ω₀)', color='blue')
plt.plot(t, response_crit, label='Críticamente Amortiguado (α = ω₀)', color='green', linestyle='--')
plt.plot(t, response_over, label='Sobreamortiguado (α > ω₀)', color='red', linestyle='-.')

# Añadir títulos y etiquetas
plt.title('Respuesta de un Circuito RLC: Subamortiguado, Críticamente Amortiguado, Sobreamortiguado')
plt.xlabel('Tiempo (s)')
plt.ylabel('Respuesta')
plt.xlim(0, 20)
plt.legend(loc='best')
plt.grid(True)
plt.show()
```

Como se observa en la gráfica, podemos notar cómo varía la respuesta en los tres casos:
- **Subamortiguado**: Oscilaciones amortiguadas.
- **Críticamente Amortiguado**: Decaimiento rápido sin oscilaciones.
- **Sobreamortiguado**: Decaimiento lento sin oscilaciones.


## Conclusión

Un circuito **RLC** puede comportarse bajo diferentes tipos de amortiguamiento. Mediante el ajuste de los valores de $ \alpha $ y $ \omega_0 $, es posible observar tres comportamientos distintos: **subamortiguado**, **críticamente amortiguado**, y **sobreamortiguado**. Las gráficas generadas te permiten visualizar claramente las diferencias en las respuestas del circuito.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función para los diferentes tipos de amortiguamiento
def subamortiguado(t, alpha, omega_0):
    """ Solución para el caso subamortiguado (alpha < omega_0) """
    omega_d = np.sqrt(omega_0**2 - alpha**2)
    return np.exp(-alpha * t) * (np.cos(omega_d * t) + (alpha / omega_d) * np.sin(omega_d * t))

def criticamente_amortiguado(t, alpha, omega_0):
    """ Solución para el caso críticamente amortiguado (alpha = omega_0) """
    return (1 + alpha * t) * np.exp(-alpha * t)

def sobreamortiguado(t, alpha, omega_0):
    """ Solución para el caso sobreamortiguado (alpha > omega_0) """
    s1 = -alpha + np.sqrt(alpha**2 - omega_0**2)
    s2 = -alpha - np.sqrt(alpha**2 - omega_0**2)
    return np.exp(s1 * t) + np.exp(s2 * t)

# Parámetros de entrada: alpha y omega_0
alpha = float(input("Ingrese el valor de α (amortiguamiento): "))
omega_0 = float(input("Ingrese el valor de ω₀ (frecuencia natural): "))

# Generar el tiempo para la gráfica
t = np.linspace(0, 20, 1000)

# Determinar el caso y calcular la respuesta correspondiente
if alpha < omega_0:
    response = subamortiguado(t, alpha, omega_0)
    label = 'Subamortiguado (α < ω₀)'
    color = 'blue'
elif alpha == omega_0:
    response = criticamente_amortiguado(t, alpha, omega_0)
    label = 'Críticamente Amortiguado (α = ω₀)'
    color = 'green'
else:
    response = sobreamortiguado(t, alpha, omega_0)
    label = 'Sobreamortiguado (α > ω₀)'
    color = 'red'

# Graficar la respuesta correspondiente
plt.figure(figsize=(10, 6))
plt.plot(t, response, label=label, color=color)

# Añadir títulos y etiquetas
plt.title(f'Respuesta de un Circuito RLC: {label}')
plt.xlabel('Tiempo (s)')
plt.ylabel('Respuesta')
plt.xlim(0, 20)
plt.legend(loc='best')
plt.grid(True)
plt.show()

# Instrucciones:
# 1. Al ejecutar el código, ingrese el valor de α y ω₀ cuando se soliciten.
# 2. El programa determinará automáticamente si el sistema es subamortiguado, críticamente amortiguado o sobreamortiguado.
# 3. Se graficará únicamente la respuesta correspondiente al caso detectado.
```
