# **Introducción al Factor de Potencia y su Corrección**
### Autor: ***Felipe Meza-Obando***

El **factor de potencia (FP)** es una medida que indica la eficiencia con la que la energía eléctrica se utiliza en un sistema. Se define como el **cociente entre la potencia activa (P)**, que realiza trabajo útil, y la **potencia aparente (S)**, que es la combinación de la potencia activa y la **potencia reactiva (Q)**. La potencia reactiva es aquella que no realiza trabajo útil, pero es necesaria para mantener los campos magnéticos en dispositivos como motores e inductores.

Matemáticamente, el factor de potencia se expresa como:

$
FP = \frac{P}{S} = \cos(\theta)
$

donde $ \theta $ es el ángulo de fase entre el voltaje y la corriente. Un **FP de 1 (o 100%)** indica que toda la energía suministrada se utiliza eficientemente para realizar trabajo. Sin embargo, en la práctica, muchos sistemas tienen un **factor de potencia menor a 1**, lo que implica que parte de la energía se pierde en forma de potencia reactiva.

---

### **¿Por qué es Necesaria la Corrección del Factor de Potencia?**

Un **bajo factor de potencia** en un sistema puede tener varias consecuencias negativas:
1. **Mayor demanda de corriente**: La misma potencia activa requiere más corriente para ser suministrada, lo que puede sobrecargar los conductores.
2. **Pérdidas adicionales**: El exceso de corriente provoca mayores pérdidas por efecto Joule (\(I^2R\)) en las líneas de transmisión y distribución.
3. **Penalizaciones por parte de la compañía eléctrica**: Muchas empresas de suministro eléctrico imponen cargos adicionales si el factor de potencia cae por debajo de ciertos niveles.

---

### **Corrección del Factor de Potencia**

La **corrección del factor de potencia** consiste en añadir **capacitores** u otros dispositivos al sistema eléctrico. Los capacitores proporcionan potencia reactiva negativa (capacitiva) que **compensa la potencia reactiva inductiva** generada por equipos como motores, mejorando así el factor de potencia.

El objetivo principal es:
- **Reducir el ángulo de desfase** entre el voltaje y la corriente.
- **Optimizar el uso de energía**, minimizando la corriente necesaria para una determinada potencia activa.
- **Evitar sobrecargas** en las líneas y **reducir las pérdidas** en el sistema.

---


### **Impacto del Factor de Potencia Adelantado o Atrasado**

El **tipo de factor de potencia** (adelantado o atrasado) juega un papel importante en los sistemas eléctricos, ya que determina si la potencia reactiva es **inductiva** o **capacitiva**. Esta diferencia influye directamente en el tipo de componente utilizado para la corrección (capacitor o inductor) y en cómo se realiza el cálculo de la **compensación de la potencia reactiva**.

---

### **Factor de Potencia Atrasado**
- Un **factor de potencia atrasado** ocurre cuando la **corriente se retrasa** respecto al voltaje.
- Esto es típico en sistemas que contienen **cargas inductivas**, como motores, transformadores e inductores.
- En este caso, se genera **potencia reactiva inductiva (Q positiva)**, lo que indica que la corriente está siendo utilizada para crear campos magnéticos.
- **Corrección**: Se utiliza un **capacitor** para **compensar la potencia reactiva inductiva**, ya que los capacitores proporcionan potencia reactiva negativa (capacitiva).

**Fórmula relevante**:

$
Q_{correctiva} = Q_{inicial} - Q_{deseada} \quad \text{(Compensación capacitiva)}
$

---

### **Factor de Potencia Adelantado**
- Un **factor de potencia adelantado** ocurre cuando la **corriente se adelanta** respecto al voltaje.
- Esto es típico en sistemas con **cargas capacitivas**, como bancos de capacitores o algunas fuentes de energía.
- En este caso, se genera **potencia reactiva negativa (Q negativa)**, lo que indica un exceso de carga capacitiva.
- **Corrección**: Se puede utilizar un **inductor** para **compensar la potencia reactiva capacitiva**, ya que los inductores generan potencia reactiva positiva.

**Fórmula relevante**:

$
Q_{correctiva} = Q_{deseada} - Q_{inicial} \quad \text{(Compensación inductiva)}
$





### **Ejemplo Práctico (FP ATRASO)**

Calcularemos la capacitancia necesaria para **corregir el factor de potencia** de un sistema que contiene un motor. También visualizaremos cómo cambia el **triángulo de potencia** antes y después de la corrección, ilustrando el impacto positivo de esta práctica.


## **Caso de Corrección del Factor de Potencia**

- Considere un motor industrial de $50 \, kW$ y $FP = 0.8$ de atraso que se encuentra conectado a una tensión de $230 \, V_{rms}$ y $60 \, Hz$.

- Se desea elevar el factor de potencia a $0.95$ de atraso.

- Describa el procedimiento y calcule el valor de la **capacitancia** que se necesita para lograr el objetivo.



### **Procedimiento**

1. **Determinar la potencia compleja original** $\overline{S}_{inicial}$.

2. **Determinar la potencia compleja deseada** $\overline{S}_{nueva}$.

3. **Encontrar la potencia absorbida por la carga correctiva**.

4. **Determinar la impedancia necesaria para la corrección**.

5. **Calcular el valor del componente requerido**.




### **Desarrollo**

#### **Potencia Compleja Original**

- Se tiene $FP = 0.8$ en atraso  
  $ \Rightarrow \theta_u - \phi_i = \alpha = \cos^{-1}(0.8) = 36.87^\circ $  
  $ \Rightarrow \theta_{inicial} = 36.87^\circ $

- Como $FP = \frac{P}{S}$, con $S = |\overline{S}|$:

  $
  |\overline{S}| = \frac{50 \, kW}{0.8} = 62.5 \, kVA
  \quad \text{y} \quad \overline{S} = |\overline{S}| \angle \alpha
  $

- Por lo tanto:

  $
  \overline{S}_{inicial} = 62.5 \angle 36.87^\circ \, [kVA]
  \Rightarrow 50 + j \, 37.50 \, kVA
  $

_El triángulo con las componentes de la potencia compleja serian: 50 kW de potencia activa, 37.5 kVAR de potencia reactiva, y 62.5 kVA de potencia aparente, con un ángulo de 36.87°._


### **Potencia Deseada**

- $FP_{deseado} = 0.95$ de atraso  

  $
  \Rightarrow \overline{S}_{nueva} = \frac{50 \, kW}{0.95} \angle \left[\cos^{-1}(0.95)\right]
  $

  $
  \Rightarrow \overline{S}_{nueva} = 52.63 \angle 18.19^\circ \, [kVA]
  $

  $
  \overline{S}_{nueva} = 50 + j \, 16.43 \, kVA
  $

---

### **Potencia Absorbida por la Carga Correctiva**

$
\overline{S}_{nueva} - \overline{S}_{inicial} = -j \, 21.07 \, kVAR
$

$
\Rightarrow \text{Se deben añadir } 21.07 \, kVAR \, \text{capacitivos.}
$


### **Impedancia para la Corrección**

Se conoce que:

$
\overline{S} = \frac{V_{rms}^2}{Z_c^*}
$

Entonces:

$
-j \, 21.07 \, kVAR = \frac{(230)^2}{Z_c^*}
\Rightarrow Z_c^* = \frac{(230)^2}{-j \, 21.07 \, kVAR}
$

$
Z_c^* = j \, 2.511 \, \Omega \quad \Rightarrow \quad Z_c = -j \, 2.511 \, \Omega
$

---

### **Valor del Capacitor**

Considerando que en una red industrial \(f = 60 \, Hz\), se tiene:

$
-j \, 2.511 \, \Omega = \frac{1}{j \omega C} \quad \Rightarrow \quad 2.511 \, \Omega = \frac{1}{\omega C}
\quad \text{con} \quad \omega = 2 \pi f
$

Por lo tanto:

$
C = \frac{1}{2 \pi (60)(2.511)}
\Rightarrow C = 1056 \, \mu F
$


Con esto se determina la capacitancia necesaria para lograr la corrección del factor de potencia a los valores deseados.


### **Código Python: Cálculo Automático de Capacitancia y Triángulo de Potencia**



```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("seaborn-v0_8-paper")

def calcular_capacitancia(potencia_activa, fp_inicial, fp_deseado, voltaje_rms, frecuencia):
    """
    Calcula la capacitancia necesaria para corregir el factor de potencia.
    """
    # Ángulos de fase inicial y deseada
    theta_inicial = np.arccos(fp_inicial)
    theta_deseado = np.arccos(fp_deseado)

    # Potencias reactivas inicial y deseada
    q_inicial = potencia_activa * np.tan(theta_inicial)  # kVAR
    q_deseada = potencia_activa * np.tan(theta_deseado)  # kVAR

    # Potencia reactiva a corregir
    q_correctiva = q_inicial - q_deseada

    # Impedancia capacitiva necesaria
    zc = (voltaje_rms ** 2) / (q_correctiva * 1000)  # Ohmios

    # Capacitancia (C) en microfaradios
    omega = 2 * np.pi * frecuencia
    capacitancia = 1 / (omega * zc) * 1e6  # Convertir a µF

    return capacitancia, q_inicial, q_deseada, q_correctiva

def graficar_triangulo_potencia(potencia_activa, q_inicial, q_deseada):
    """
    Dibuja el triángulo de potencia antes y después de la corrección con estilo mejorado.
    """
    # Crear figura
    plt.figure(figsize=(8, 6))

    # Dibujar potencias activa y reactiva inicial
    plt.plot([0, potencia_activa], [0, 0], color='blue', linewidth=2)  # P activa
    plt.plot([potencia_activa, potencia_activa], [0, q_inicial], color='purple', linewidth=2,
             label=f'{q_inicial:.2f} kVAR (Inicial)')

    # Dibujar potencia reactiva deseada
    plt.plot([potencia_activa, potencia_activa], [0, q_deseada], color='orange', linewidth=2,
             label=f'{q_deseada:.2f} kVAR (Deseada)')

    # Dibujar hipotenusas (potencia aparente)
    plt.plot([0, potencia_activa], [0, q_inicial], color='green', linestyle='-', linewidth=2,
             label=f'{np.sqrt(potencia_activa**2 + q_inicial**2):.2f} kVA (Inicial)')
    plt.plot([0, potencia_activa], [0, q_deseada], color='darkgreen', linestyle='--', linewidth=2,
             label=f'{np.sqrt(potencia_activa**2 + q_deseada**2):.2f} kVA (Corregida)')

    # Etiquetas de los ejes y título
    plt.xlabel('Potencia Activa (kW)')
    plt.ylabel('Potencia Reactiva (kVAR)')
    plt.title('Triángulo de Potencia: Antes y Después de la Corrección')

    # Leyenda dinámica
    plt.legend(loc='upper left')

    # Ajustes del gráfico
    plt.grid(True)
    plt.xlim(0, potencia_activa + 10)
    plt.ylim(0, max(q_inicial, q_deseada) + 10)

    plt.show()

# Parámetros del problema
potencia_activa = 40  # kW
fp_inicial = 0.7
fp_deseado = 0.99
voltaje_rms = 230  # Voltios RMS
frecuencia = 60  # Hz

# Calcular capacitancia y graficar el triángulo de potencia
capacitancia, q_inicial, q_deseada, q_correctiva = calcular_capacitancia(
    potencia_activa, fp_inicial, fp_deseado, voltaje_rms, frecuencia)

print(f"Capacitancia requerida: {capacitancia:.2f} µF")
print(f"Potencia reactiva inicial: {q_inicial:.2f} kVAR")
print(f"Potencia reactiva deseada: {q_deseada:.2f} kVAR")
print(f"Potencia correctiva necesaria: {q_correctiva:.2f} kVAR")

# Generar el gráfico
graficar_triangulo_potencia(potencia_activa, q_inicial, q_deseada)
```


### **Descripción del Código**

1. **Función `calcular_capacitancia()`**:
   - Calcula la **capacitancia necesaria** para la corrección del factor de potencia.
   - Retorna también las potencias reactivas antes y después de la corrección.

2. **Función `graficar_triangulo_potencia()`**:
   - Dibuja el **triángulo de potencia** mostrando las potencias activa y reactiva inicial y corregida.

3. **Parámetros Ejemplo del Problema**:
   - **Potencia activa**: 50 kW.
   - **Voltaje RMS**: 230 V.
   - **Frecuencia**: 60 Hz.
   - **Factor de potencia inicial**: 0.8.
   - **Factor de potencia deseado**: 0.95.

---

### **Ejemplo de Salida**

- Capacitancia requerida: 1056 µF.
- Potencia reactiva inicial: 37.5 kVAR.
- Potencia reactiva deseada: 16.43 kVAR.
- Potencia correctiva necesaria: 21.07 kVAR.

El gráfico mostrará el **triángulo de potencia** antes y después de la corrección, ilustrando claramente la reducción en la potencia reactiva.

### **Ejemplo Práctico (FP ADELANTO)**

En este caso, la corrección se realiza mediante inductores, ya que generan potencia reactiva positiva para compensar el exceso de potencia reactiva capacitiva.

```python
import numpy as np
import matplotlib.pyplot as plt

def calcular_inductancia(potencia_activa, fp_inicial, fp_deseado, voltaje_rms, frecuencia):
    """
    Calcula la inductancia necesaria para corregir el factor de potencia adelantado.
    """
    # Ángulos de fase inicial y deseada
    theta_inicial = np.arccos(fp_inicial)
    theta_deseado = np.arccos(fp_deseado)

    # Potencias reactivas inicial y deseada
    q_inicial = -potencia_activa * np.tan(theta_inicial)  # kVAR (negativa por adelanto)
    q_deseada = -potencia_activa * np.tan(theta_deseado)  # kVAR

    # Potencia reactiva a corregir (necesitamos generar Q positiva)
    q_correctiva = abs(q_deseada - q_inicial)  # Siempre positiva

    # Impedancia inductiva necesaria
    zl = (voltaje_rms ** 2) / (q_correctiva * 1000)  # Ohmios

    # Inductancia (L) en Henrios
    omega = 2 * np.pi * frecuencia
    inductancia = zl / omega  # Convertir a Henrios

    return inductancia, q_inicial, q_deseada, q_correctiva

def graficar_triangulo_potencia_adelantado(potencia_activa, q_inicial, q_deseada):
    """
    Dibuja el triángulo de potencia para el caso de FP adelantado con corrección.
    """
    # Crear figura
    plt.figure(figsize=(8, 6))

    # Dibujar potencias activa y reactiva inicial
    plt.plot([0, potencia_activa], [0, 0], color='blue', linewidth=2)  # P activa
    plt.plot([potencia_activa, potencia_activa], [0, q_inicial], color='purple', linewidth=2,
             label=f'{q_inicial:.2f} kVAR (Inicial)')

    # Dibujar potencia reactiva deseada
    plt.plot([potencia_activa, potencia_activa], [0, q_deseada], color='orange', linewidth=2,
             label=f'{q_deseada:.2f} kVAR (Deseada)')

    # Dibujar hipotenusas (potencia aparente)
    plt.plot([0, potencia_activa], [0, q_inicial], color='green', linestyle='-', linewidth=2,
             label=f'{np.sqrt(potencia_activa**2 + q_inicial**2):.2f} kVA (Inicial)')
    plt.plot([0, potencia_activa], [0, q_deseada], color='darkgreen', linestyle='--', linewidth=2,
             label=f'{np.sqrt(potencia_activa**2 + q_deseada**2):.2f} kVA (Corregida)')

    # Etiquetas de los ejes y título
    plt.xlabel('Potencia Activa (kW)')
    plt.ylabel('Potencia Reactiva (kVAR)')
    plt.title('Triángulo de Potencia: FP Adelantado y Corrección')

    # Leyenda dinámica
    plt.legend(loc='upper left')

    # Ajustes del gráfico
    plt.grid(True)
    plt.xlim(0, potencia_activa + 10)
    plt.ylim(min(q_inicial, q_deseada) - 10, 10)

    plt.show()

# Parámetros del problema
potencia_activa = 50  # kW
fp_inicial = 0.95  # FP adelantado (inicial)
fp_deseado = 0.8  # FP deseado (más inductivo)
voltaje_rms = 230  # Voltios RMS
frecuencia = 60  # Hz

# Calcular inductancia y graficar el triángulo de potencia
inductancia, q_inicial, q_deseada, q_correctiva = calcular_inductancia(
    potencia_activa, fp_inicial, fp_deseado, voltaje_rms, frecuencia
)

print(f"Inductancia requerida: {inductancia:.2e} H")
print(f"Potencia reactiva inicial: {q_inicial:.2f} kVAR")
print(f"Potencia reactiva deseada: {q_deseada:.2f} kVAR")
print(f"Potencia correctiva necesaria: {q_correctiva:.2f} kVAR")

# Generar el gráfico
graficar_triangulo_potencia_adelantado(potencia_activa, q_inicial, q_deseada)

```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("seaborn-v0_8-paper")

def calcular_capacitancia(potencia_activa, fp_inicial, fp_deseado, voltaje_rms, frecuencia):
    """
    Calcula la capacitancia necesaria para corregir el factor de potencia.
    """
    # Ángulos de fase inicial y deseada
    theta_inicial = np.arccos(fp_inicial)
    theta_deseado = np.arccos(fp_deseado)

    # Potencias reactivas inicial y deseada
    q_inicial = potencia_activa * np.tan(theta_inicial)  # kVAR
    q_deseada = potencia_activa * np.tan(theta_deseado)  # kVAR

    # Potencia reactiva a corregir
    q_correctiva = q_inicial - q_deseada

    # Impedancia capacitiva necesaria
    zc = (voltaje_rms ** 2) / (q_correctiva * 1000)  # Ohmios

    # Capacitancia (C) en microfaradios
    omega = 2 * np.pi * frecuencia
    capacitancia = 1 / (omega * zc) * 1e6  # Convertir a µF

    return capacitancia, q_inicial, q_deseada, q_correctiva

def graficar_triangulo_potencia(potencia_activa, q_inicial, q_deseada):
    """
    Dibuja el triángulo de potencia antes y después de la corrección con estilo mejorado.
    """
    # Crear figura
    plt.figure(figsize=(8, 6))

    # Dibujar potencias activa y reactiva inicial
    plt.plot([0, potencia_activa], [0, 0], color='blue', linewidth=2)  # P activa
    plt.plot([potencia_activa, potencia_activa], [0, q_inicial], color='purple', linewidth=2,
             label=f'{q_inicial:.2f} kVAR (Inicial)')

    # Dibujar potencia reactiva deseada
    plt.plot([potencia_activa, potencia_activa], [0, q_deseada], color='orange', linewidth=2,
             label=f'{q_deseada:.2f} kVAR (Deseada)')

    # Dibujar hipotenusas (potencia aparente)
    plt.plot([0, potencia_activa], [0, q_inicial], color='green', linestyle='-', linewidth=2,
             label=f'{np.sqrt(potencia_activa**2 + q_inicial**2):.2f} kVA (Inicial)')
    plt.plot([0, potencia_activa], [0, q_deseada], color='darkgreen', linestyle='--', linewidth=2,
             label=f'{np.sqrt(potencia_activa**2 + q_deseada**2):.2f} kVA (Corregida)')

    # Etiquetas de los ejes y título
    plt.xlabel('Potencia Activa (kW)')
    plt.ylabel('Potencia Reactiva (kVAR)')
    plt.title('Triángulo de Potencia: Antes y Después de la Corrección')

    # Leyenda dinámica
    plt.legend(loc='upper left')

    # Ajustes del gráfico
    plt.grid(True)
    plt.xlim(0, potencia_activa + 10)
    plt.ylim(0, max(q_inicial, q_deseada) + 10)

    plt.show()

# Solicitar valores al usuario
try:
    potencia_activa = float(input("Ingrese la potencia activa (kW): "))
    fp_inicial = float(input("Ingrese el factor de potencia inicial (0-1): "))
    fp_deseado = float(input("Ingrese el factor de potencia deseado (0-1): "))
    voltaje_rms = float(input("Ingrese el voltaje RMS (V): "))
    frecuencia = float(input("Ingrese la frecuencia (Hz): "))

    if not (0 < fp_inicial <= 1 and 0 < fp_deseado <= 1):
        raise ValueError("El factor de potencia debe estar entre 0 y 1.")

    if fp_deseado < fp_inicial:
        raise ValueError("El factor de potencia deseado debe ser mayor que el inicial.")

    # Calcular capacitancia y graficar el triángulo de potencia
    capacitancia, q_inicial, q_deseada, q_correctiva = calcular_capacitancia(
        potencia_activa, fp_inicial, fp_deseado, voltaje_rms, frecuencia)

    # Mostrar resultados
    print(f"\nResultados de la corrección del factor de potencia:\n")
    print(f"Capacitancia requerida: {capacitancia:.2f} µF")
    print(f"Potencia reactiva inicial: {q_inicial:.2f} kVAR")
    print(f"Potencia reactiva deseada: {q_deseada:.2f} kVAR")
    print(f"Potencia correctiva necesaria: {q_correctiva:.2f} kVAR")

    # Generar el gráfico
    graficar_triangulo_potencia(potencia_activa, q_inicial, q_deseada)

except ValueError as e:
    print(f"Error en la entrada de datos: {e}")
```
