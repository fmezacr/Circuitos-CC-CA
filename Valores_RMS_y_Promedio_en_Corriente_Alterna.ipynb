## **2. Definición del Valor Promedio**
El **valor promedio** de una señal periódica se define como:

$
V_{avg} = \frac{1}{T} \int_0^T v(t) dt
$

Donde:
- $ V_{avg} $ es el valor promedio de la señal.
- $ T $ es el período de la señal.
- $ v(t) $ es la función de voltaje en el tiempo.

Para una onda sinusoidal pura:

$
V_{avg} = \frac{2A}{\pi}
$

Siendo $ A $ la amplitud de la señal.

El valor promedio es útil en sistemas donde se busca medir el desplazamiento neto de una señal en un intervalo de tiempo.

#### **Ejemplos de Aplicación**
1. **Fuentes de alimentación rectificadas**: En circuitos rectificadores, el valor promedio nos permite conocer la cantidad de voltaje DC que se obtiene después de filtrar una señal CA.
2. **Sensores de corriente y voltaje**: Algunos dispositivos de medición dependen del valor promedio para estimar la energía transferida.
3. **Sistemas de control de motores**: En algunos sistemas de control de motores eléctricos, el valor promedio del voltaje aplicado afecta el rendimiento del motor.


### **Ejemplo en Python: Cálculo del Valor Promedio**

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la señal
A = 10  # Amplitud de la señal
T = 2*np.pi  # Período normalizado para seno (en radianes)
t = np.linspace(0, T, 1000)
v_t = A * np.sin(t)

# Cálculo del valor promedio
V_avg = (1/T) * np.trapz(v_t, t)

# Graficar
plt.figure(figsize=(8,4))
plt.plot(t, v_t, label='Onda Sinusoidal')
plt.axhline(V_avg, color='r', linestyle='--', label=f'Valor Promedio = {V_avg:.2f} V')
plt.title('Cálculo del Valor Promedio de una Onda Sinusoidal')
plt.xlabel('Tiempo')
plt.ylabel('Voltaje')
plt.legend()
plt.grid()
plt.show()
```

Para una onda sinusoidal pura, el valor promedio es **cero** porque la onda tiene la misma cantidad de área positiva que negativa.


## **3. Definición del Valor RMS**
El **valor RMS (Root Mean Square)** representa la potencia efectiva de una señal y se define como:

$
V_{rms} = \sqrt{\frac{1}{T} \int_0^T v^2(t) dt}
$

Para una onda sinusoidal pura:

$
V_{rms} = \frac{A}{\sqrt{2}}
$

El valor RMS es crucial porque en circuitos CA, una señal con valor RMS $ V_{rms} $ entrega la misma potencia a una resistencia que un voltaje de corriente continua de magnitud $ V_{rms} $.

El valor RMS es fundamental en ingeniería eléctrica porque determina el **efecto de calentamiento** de una señal en una carga resistiva.

#### **Ejemplos de Aplicación**
1. **Consumo de energía eléctrica**: Los medidores de electricidad en hogares y empresas utilizan valores RMS para medir la energía consumida.
2. **Diseño de electrodomésticos**: Los dispositivos eléctricos están diseñados para operar con voltajes RMS específicos, como los 120V o 230V RMS de las redes eléctricas.
3. **Equipos de audio**: La potencia de salida de amplificadores y bocinas se expresa en términos de valores RMS para reflejar la potencia real entregada al sistema de sonido.


### **Ejemplo en Python: Cálculo del Valor RMS**


```python

# Cálculo del valor RMS
V_rms = np.sqrt((1/T) * np.trapz(v_t**2, t))

# Graficar
plt.figure(figsize=(8,4))
plt.plot(t, v_t, label='Onda Sinusoidal')
plt.axhline(V_rms, color='g', linestyle='--', label=f'Valor RMS = {V_rms:.2f} V')
plt.title('Cálculo del Valor RMS de una Onda Sinusoidal')
plt.xlabel('Tiempo')
plt.ylabel('Voltaje')
plt.legend()
plt.grid()
plt.show()

```

Para una onda sinusoidal de **10V de amplitud**, el valor RMS es **7.07V**.

## **4. Comparación de Valores RMS y Promedio en Diferentes Ondas**

| Tipo de Onda        | Valor Promedio $ V_{avg} $ | Valor RMS $ V_{rms} $ |
|---------------------|----------------|----------------|
| Sinusoidal         | $ 0 $         | $ A/\sqrt{2} $ |
| Cuadrada           | $ A $         | $ A $ |
| Triangular         | $ A/2 $       | $ A/\sqrt{3} $ |


Para una onda cuadrada, los valores **RMS y promedio son iguales**, mientras que para una onda triangular, el valor RMS es menor que en la onda sinusoidal.

---

## **5. Conclusión**
- El valor **promedio** mide el voltaje promedio en un ciclo, pero en ondas simétricas como la sinusoidal es **cero**.
- El valor **RMS** representa el valor equivalente en potencia y es más útil en circuitos eléctricos.
- Diferentes tipos de ondas tienen diferentes relaciones entre su valor RMS y su amplitud.
- Los valores RMS son utilizados en la industria eléctrica para el cálculo de energía consumida y diseño de equipos eléctricos.

Cuando se especifica una tensión o corriente senoidal, a menudo se hace en términos de su valor máximo (o pico) o de su valor rms, ya que su valor promedio es de cero. Las industrias relacionadas con la potencia especifi can magnitudes fasoriales en términos de sus valores rms más que de sus valores pico. Por ejemplo, los 110 V disponibles en todos los hogares son el valor **rms** de la tensión procedente de la compañía suministradora de energía eléctrica. En análisis de potencia es conveniente expresar la tensión y la corriente en sus valores rms. Asimismo, los voltímetros y amperímetros analógicos están diseñados para leer en forma directa el valor rms de la tensión y la corriente, respectivamente.

Este conocimiento es fundamental en el análisis de circuitos de corriente alterna (CA) y permite diseñar sistemas más eficientes y seguros.

## **6. Calcula los valores RMS o promedio para la señal de tu interés**
Ahora que conoces los conceptos teóricos y sus aplicaciones, puedes experimentar con diferentes señales y calcular sus valores RMS y promedio.

A continuación, se presenta un código interactivo donde puedes ingresar tu propia función de señal y obtener los resultados:

```python
import numpy as np
import matplotlib.pyplot as plt

def calcular_rms_promedio(signal_func, T, A, num_puntos=1000):
    t = np.linspace(0, T, num_puntos)
    v_t = signal_func(t, A)

    # Cálculo de valores
    V_avg = (1/T) * np.trapz(v_t, t)
    V_rms = np.sqrt((1/T) * np.trapz(v_t**2, t))

    # Graficar
    plt.figure(figsize=(8,4))
    plt.plot(t, v_t, label='Señal ingresada')
    plt.axhline(V_avg, color='r', linestyle='--', label=f'Valor Promedio = {V_avg:.2f} V')
    plt.axhline(V_rms, color='g', linestyle='--', label=f'Valor RMS = {V_rms:.2f} V')
    plt.title('Cálculo del Valor RMS y Promedio')
    plt.xlabel('Tiempo')
    plt.ylabel('Voltaje')
    plt.legend()
    plt.grid()
    plt.show()

    return V_avg, V_rms

# Ejemplo de uso: prueba con una onda sinusoidal
def onda_seno(t, A):
    return A * np.sin(t)

T = 2*np.pi  # Período
A = 10  # Amplitud
V_avg, V_rms = calcular_rms_promedio(onda_seno, T, A)
print(f"Valor Promedio: {V_avg:.2f} V")
print(f"Valor RMS: {V_rms:.2f} V")
```
